*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
````

````python
df1 = df.groupby(by=ContinentDict,axis=0).size().to_frame(name='size')
````

````python
df2 = df.groupby(by=ContinentDict,axis=0).mean()
````

````python
df2 = df2.iloc[:,10:20].mean(axis = 1).to_frame(name='mean')
````

````python
df3 = df.groupby(by=ContinentDict,axis=0).sum()
````

````python
df3 = df3.iloc[:,10:20].mean(axis = 1).to_frame(name='sum')
````

````python
df4 = df.groupby(by=ContinentDict,axis=0).std()
````

````python
df4 = df4.iloc[:,10:20].mean(axis = 1).to_frame(name='std')
````

````python
df_final = pd.merge(pd.merge(pd.merge(df1, df2, left_index=True, right_index=True),df3,left_index=True, right_index=True),df4,left_index=True, right_index=True)
````
